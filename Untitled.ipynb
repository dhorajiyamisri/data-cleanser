```python
Missing Value Imputation and Outlier Detection
```

```python
import pandas as pd
import numpy as np

# generate patient dataset

np.random.seed(42)
df = pd.DataFrame({
    "patient_id":[f"P{i}" for i in range(1,101)],
    "age":np.random.randint(20,80,100),
    "gender":np.random.choice(["Male","Female"],100),
    "region":np.random.choice(["North","South","East","West"],100),
    "bmi":np.random.normal(25,5,100),
    "blood_pressure":np.random.normal(120,15,100),
    "cholesterol":np.random.normal(200,30,100),
    "glucose":np.random.normal(100,20,100),
    "disease_risk":np.random.choice([0,1],100)
})
df.to_csv("patient_health_records.csv",index=False)
```

```python
df.head()
```

```python
# introduce missing value

df.loc[0:5,"age"] = np.nan
df.loc[10:15,"gender"] = np.nan
df.loc[20:25,"region"] = np.nan
df.loc[30:35,"bmi"] = np.nan
df.loc[40:45,"cholesterol"] = np.nan
df.loc[50:55,"glucose"] = np.nan
print(df)
```

```python
# initial data exploration

# dataset information
df.info()
```

```python
# Statistical Summary
df.describe()
```

```python
# Missing Values Summary
df.isnull().sum()
```

```python
# Missing Value Imputation
# Method 1: Simple Imputer (Mean)

from sklearn.impute import SimpleImputer

mean_imp = SimpleImputer(strategy="mean")

df["bmi"] = mean_imp.fit_transform(df[["bmi"]]).ravel()

df["bmi"].isnull().sum()
```

```python
# Method 2: Most Frequent Imputer

cat_imp = SimpleImputer(
    strategy="most_frequent"
)

df["gender"] = cat_imp.fit_transform(
    df[["gender"]]
).ravel()

df["region"] = cat_imp.fit_transform(
    df[["region"]]
).ravel()

print(df[["gender","region"]].isnull().sum())
```

```python
# Method 3: Missing Indicator + Random Sample

df["age_missing"] = df["age"].isnull().astype(int)

sample = df["age"].dropna().sample(
    df["age"].isnull().sum(),
    random_state=42
)

df.loc[df["age"].isnull(),"age"] = sample.values

df.head()
```

```python
# Method 4: KNN Imputer
from sklearn.impute import KNNImputer

knn = KNNImputer()

cols = [
    "age",
    "bmi",
    "blood_pressure",
    "cholesterol",
    "glucose"
]

df[cols] = knn.fit_transform(df[cols])
print(df)
```

```python
# Method 5: MICE
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer

mice = IterativeImputer(random_state=42)
df[cols] = mice.fit_transform(df[cols])
print(df)
```

```python
# Outlier Detection
# Z-Score Method
from scipy.stats import zscore

df["zscore_bmi"] = zscore(df["bmi"])

outliers = df[
    abs(df["zscore_bmi"]) > 3
]

print("Outliers:",len(outliers))
```

```python
# IQR Method
Q1 = df["bmi"].quantile(0.25)
Q3 = df["bmi"].quantile(0.75)

IQR = Q3 - Q1

lower = Q1 - 1.5*IQR
upper = Q3 + 1.5*IQR

print(lower, upper)
```

```python
# Percentile Method
print(df["bmi"].quantile(0.01))
print(df["bmi"].quantile(0.99))
```

```python
# Outlier Treatment
from scipy.stats.mstats import winsorize

df["bmi"] = winsorize(
    df["bmi"],
    limits=[0.01,0.01]
)

df.head()
```

```python
# Final Dataset Validation
df.isnull().sum()
```

```python
df.to_csv(
    "final_clean_patient_data.csv",
    index=False
)

print("Final Dataset Saved Successfully")
```
